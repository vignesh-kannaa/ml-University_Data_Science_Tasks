```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
ds=pd.read_excel('ENB2012_data.xlsx')
ds
```

```python
df = ds.rename({'X1': 'Relative Compactness', 'X2': 'Surface Area', 'X3': 'Wall Area', 'X4': 'Roof Area', 'X5': 'Overall Height', 'X6': 'Orientation', 'X7': 'Glazing Area', 'X8': 'Glazing Area Distribution', 'Y1': 'Heating Load', 'Y2': 'Cooling Load'}, axis=1)  # new method
```

```python
dia=pd.plotting.scatter_matrix(df,figsize=(20,20))
```

<h2>Regression for X1,X2,Y1</h2>

```python
X=ds.iloc[:,0:2].values
y=ds.iloc[:,-2].values
```

```python
regressor = LinearRegression()
model=regressor.fit(X, y)
y_pred = regressor.predict(X)
```

```python
coefficient=regressor.coef_
print('co-efficient:',coefficient)
mse = mean_squared_error(y, y_pred)
print('mean Squared error:',mse)
cod=regressor.score(X, y);
print('co-efficient of determination:',cod)
```
